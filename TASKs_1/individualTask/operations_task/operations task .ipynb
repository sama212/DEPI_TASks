# Logical operation



**can use with a boolean values  or  numbers**
<pre>     but when use it with numbers, there are a few things to note:

            1. [0] represent False and any [non-zero number] represent True 
            2. It return a number not [True,False]:
   
                  1. With zero operator at any side it always return 0
                  2. In other cases, it return the last operator read after check truthy of expression:
                                                        
                        <span style="color:Green">With or --> it returns the first true operand it encounters,
                                    and once it finds the first true operand, the last operand is not evaluated at all.
                        With and --> it returns the second operand
                                    because it must check the truth of both sides to determine the expression’s result.</span>

<span style = "color:Red "><u>note:</u>

      The above applies to binary operations [and, or] but [not] which unary operations that work in the following way:
      
            1. with zere it print True
            2. with any non-zero number it print False

</span>
                                                              </pre>

```python
#boolean values
bool_value1 = True
bool_value2 = False

#numerical values
number_value1 = 0
number_value2 = 5
number_value3 = 4


#and operation with boolean values
print("<<and operation with boolean values>>\n")
print(f" {bool_value1:<3} and   {bool_value2:<3} result --> " , bool_value1 and bool_value2) 

print("--------------------------------------------------------------")

#and operation with numerical values
print("<<and operation with numerical values>>\n")
print(f" {number_value1:<3} and   {number_value2:<3} result --> " , number_value1 and number_value2) 
print(f" {number_value2:<3} and   {number_value3:<3} result --> " , number_value2 and number_value3) 
print(f" {number_value3:<3} and   {number_value2:<3} result --> " , number_value3 and number_value2) 

print("--------------------------------------------------------------")

#or operation with boolean values
print("<<or operation with boolean values>>\n")
print(f" {bool_value1:<3} or   {bool_value2:<3} result --> " , bool_value1 or bool_value2) 

print("--------------------------------------------------------------")

#or operation with numerical values
print("<<or operation with numerical values>>\n")
print(f" {number_value1:<3} or   {number_value2:<3} result --> " , number_value1 or number_value2) 
print(f" {number_value2:<3} or   {number_value3:<3} result --> " , number_value2 or number_value3) 
print(f" {number_value3:<3} or   {number_value2:<3} result --> " , number_value3 or number_value2) 

print("--------------------------------------------------------------")

#not operation with boolean values
print("<<not operation with boolean values>>\n")
print(f" not   {bool_value1:<3} result --> " , not bool_value1) 
print(f" not   {bool_value2:<3} result --> " , not bool_value2) 

print("--------------------------------------------------------------")

#not operation with numerical values
print("<<not operation with numerical values>>\n")
print(f" not   {number_value1:<3} result --> " , not number_value1) 
print(f" not   {number_value2:<3} result --> " , not number_value2) 
print(f" not   {number_value3:<3} result --> " , not number_value3) 
```

# Bitwise operations

```python
number1 = 3  #0011
number2 = 7  #0111
number3 = 8  #1000

# & operator
print(f" {number1:<3} &   {number2:<3} result --> " , number1 & number2)  # 0011 & 0111 --> 0011 = 3
print(f" {number2:<3} &   {number3:<3} result --> " , number2 & number3)  # 0111 & 1000 --> 0000 = 0

print("--------------------------------------------------------------")

# | operator
print(f" {number1:<3} |   {number2:<3} result --> " , number1 | number2)  # 0011 | 0111 --> 0111 = 7
print(f" {number2:<3} |   {number3:<3} result --> " , number2 | number3)  # 0111 | 1000 --> 1111 = 15

print("--------------------------------------------------------------")

# ~ operator
print(f" ~   {number1:<3} result --> " ,~ number1)  # 0000 0011 --> 1111 1100, Msb --> 1111 represent {-}  and reverse {1111 1100} to 0000 0011 + 1 = 0000 0100  --> 4 so all be -4
print(f" ~   {number2:<3} result --> " ,~ number2)  # 0000 0111 --> 1111 1000, Msb --> 1111 represent {-}  and reverse {1111 1000} to 0000 0111 + 1 = 0000 1000  --> 8 so all be -8
print(f" ~   {number3:<3} result --> " ,~ number3)  # 0000 1000 --> 1111 0111, Msb --> 1111 represent {-}  and reverse {1111 0111} to 0000 1000 + 1 = 0000 1001  --> 9 so all be -9

print("--------------------------------------------------------------")

# ^ operator
print(f" {number1:<3} ^   {number2:<3} result --> " , number1 ^ number2)  # 0011 ^ 0111 --> 0100 = 4
print(f" {number2:<3} ^   {number3:<3} result --> " , number2 ^ number3)  # 0111 ^ 1000 --> 1111 = 15

print("--------------------------------------------------------------")

#shift operation by zero 

# >> operator
print(f" ≫{number1:<3} which represent [{bin(number1)[2:]:0>4}] result --> {number1 >> 1}  , represent after right shift [{bin(number1 >> 1)[2:]:0>4}]")
print(f" ≫{number2:<3} which represent [{bin(number2)[2:]:0>4}] result --> {number2 >> 1}  , represent after right shift [{bin(number2 >> 1)[2:]:0>4}]")
print(f" ≫{number3:<3} which represent [{bin(number3)[2:]:0>4}] result --> {number3 >> 1}  , represent after right shift [{bin(number3 >> 1)[2:]:0>4}]")

print("--------------------------------------------------------------")

# ≪ operator
print(f" {number1}≪ which represent [{bin(number1)[2:]:0>4}] result --> {number1 << 1}   , represent after right shift [  {bin(number1 << 1)[2:]:0>4} ]")
print(f" {number2}≪ which represent [{bin(number2)[2:]:0>4}] result --> {number2 << 1}  , represent after right shift [  {bin(number2 << 1)[2:]:0>4} ]")
print(f" {number3}≪ which represent [{bin(number3)[2:]:0>4}] result --> {number3 << 1}  , represent after right shift [ {bin(number3 << 1)[2:]:0>4} ]")

print("--------------------------------------------------------------")
```
